```python
from xrd_decosmic.models.data_model import DataModel
from xrd_decosmic.models.params_model import ParamsModel
```

```python
fileName = "C:/Users/ctyja/Desktop/origami/origami_0000.tif"
dataModel = DataModel(fileName)
```

```python
thDonut = '15'
thMask = '0.1'
thStrike = '3'
winStrike = '3'
```

```python
paramsModel = ParamsModel(thDonut, thMask, thStrike, winStrike)
dataModel.loadParams(paramsModel)
```

```python
dataModel.avgCleanImg()
```

```python
fabio.tifimage.TifImage(dataModel.imgAvg).save("avg.tif")
fabio.tifimage.TifImage(dataModel.imgCleanAvg).save("avg_clean.tif")
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12,10))
plt.subplot(2,3,1)
plt.title('Average')
plt.imshow(dataModel.imgAvg, cmap='hot', vmin=0, vmax=0.2)
plt.axis('off')
plt.subplot(2,3,2)
plt.title('Cleaned Average')
plt.imshow(dataModel.imgCleanAvg, cmap='hot', vmin=0, vmax=0.2)
plt.axis('off')
plt.subplot(2,3,3)
plt.title('Difference')
plt.imshow(dataModel.imgDiffAvg, cmap='hot', vmin=0, vmax=0.02)
plt.axis('off')
plt.subplot(2,3,4)
plt.title('Subtracted Donut')
plt.imshow(dataModel.subDonutAvg, cmap='hot', vmin=0, vmax=0.02)
plt.axis('off')
plt.subplot(2,3,5)
plt.title('Subtracted Strike')
plt.imshow(dataModel.subStrikeAvg, cmap='hot', vmin=0, vmax=0.02)
plt.axis('off')
plt.subplot(2,3,6)
plt.title('Mask')
plt.imshow(1-dataModel.combinedMask, cmap='hot', vmin=0, vmax=1)
plt.axis('off')
plt.tight_layout()
plt.show()

```
